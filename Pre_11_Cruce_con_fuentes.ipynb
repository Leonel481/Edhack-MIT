```python
import pandas as pd
import numpy as np
from simpledbf import Dbf5
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
def leerDBF(path_):
    Dbf = Dbf5(path_, codec='cp850')
    df = Dbf.to_dataframe()
    return df
```
